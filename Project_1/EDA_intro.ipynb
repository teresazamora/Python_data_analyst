```python
import pandas as pd
import ast  
import matplotlib.pyplot as plt 
import seaborn as sns

from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()
```

```python
df.info()
```

```python
df.job_posted_date = pd.to_datetime(df['job_posted_date'])
```

```python
df.job_skills = df.job_skills.apply(lambda skill: ast.literal_eval(skill) if pd.notna(skill) else skill)
```

```python
df_DA_Italy = df[(df['job_country'] == 'Italy') & (df['job_title_short'] == 'Data Analyst')]
```

```python
df_DA_Italy.head()
```

```python
df_plot = df_DA_Italy['job_location'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)

plt.xlabel('Number of job')
plt.ylabel('')
plt.title('Count of job location for Data Analyst in Italy')
plt.show()
plt.tight_layout()
```

```python
dict_columns = {
  'job_work_from_home': 'Remote work posibility',
  'job_no_degree_mention': 'No degree requiered'
}

fig, ax = plt.subplots(1,2)

for i,(column,title) in enumerate(dict_columns.items()):
  ax[i].pie(df_DA_Italy[column].value_counts(), labels=['False', 'True'], autopct='%1.1f%%', startangle=90)
  ax[i].set_title(title)

plt.show()
plt.tight_layout()
```

```python
df_company = df_DA_Italy.company_name.value_counts().head(10).to_frame()
```
