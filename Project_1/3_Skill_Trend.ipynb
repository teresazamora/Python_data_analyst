```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
import seaborn as sns

from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()
```

```python
df.info()
```

```python
df.job_posted_date = pd.to_datetime(df.job_posted_date)
```

```python
df.job_skills = df.job_skills.apply(lambda skill: ast.literal_eval(skill) if pd.notna(skill) else skill)
```

```python
df_DA_Italy = df[(df['job_title_short'] == 'Data Analyst') & (df['job_location'] == 'Italy')]
```

```python
df_DA_Italy.head()
```

```python
df_DA_Italy['job_posted_month_nr'] = df_DA_Italy['job_posted_date'].dt.month
```

```python
df_DA_Italy.head()
```

```python
df_DA_Italy_explode = df_DA_Italy.explode('job_skills')
```

```python
df_Italy_pivot = df_DA_Italy_explode.pivot_table(index='job_posted_month_nr', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
df_Italy_pivot.loc['Total'] = df_Italy_pivot.sum()
df_Italy_pivot = df_Italy_pivot[df_Italy_pivot.loc['Total'].sort_values(ascending=False).index]
df_Italy_pivot = df_Italy_pivot.drop(['Total'])
```

```python
df_Italy_pivot
```

```python
df_DA_Italy_month = df_DA_Italy.groupby('job_posted_month_nr').size()
```

```python
df_Italy_pivot = df_Italy_pivot.div(df_DA_Italy_month/100, axis=0)
```

```python
df_Italy_pivot
```

```python
sns.lineplot(df_Italy_pivot, legend=False)
```

```python
df_Italy_pivot = df_Italy_pivot.reset_index()
df_Italy_pivot['job_posted_month_nr'] = pd.to_datetime(df_Italy_pivot['job_posted_month_nr'], format='%m').dt.strftime('%b') 
df_Italy_pivot = df_Italy_pivot.set_index('job_posted_month')
df_Italy_pivot = df_Italy_pivot.drop(columns='job_posted_month_nr')
```

```python
df_Italy_pivot = df_Italy_pivot.drop(columns='level_0')
```

```python
df_Italy_pivot
```

```python
df_Italy_pivot = df_Italy_pivot.drop(columns='index')
```

```python
df_Italy_pivot = df_Italy_pivot.reset_index()
```

```python
df_Italy_pivot = df_Italy_pivot.set_index('job_posted_month_nr')
```

```python
df_Italy_pivot
```

```python
df_Italy_pivot = df_Italy_pivot.drop(columns='index')
```

```python
df_Italy_pivot = df_Italy_pivot.iloc[:, 0:5]
```

```python
from matplotlib.ticker import PercentFormatter
sns.lineplot(df_Italy_pivot, dashes=False, legend=False)
plt.xlabel('2023')
plt.ylabel('Job Skills')
plt.title('Trending top 5 skills for Data Analyst in Italy')
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))
sns.despine()

for i in range(5):
  if i == 1:
    plt.text(11.2, df_Italy_pivot.iloc[-1, i] + 1.4, df_Italy_pivot.columns[i], rotation= 30.)
  else:  
    plt.text(11.2, df_Italy_pivot.iloc[-1, i], df_Italy_pivot.columns[i], rotation=30.)

plt.show()
```
