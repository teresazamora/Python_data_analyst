```python
import pandas as pd
import matplotlib.pyplot as plt


from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()
```

```python
df
```

```python
df.info()
```

```python
df.job_posted_date = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
top_country = df.job_country.value_counts().head(6).index
df_job_country_salary = df.pivot_table(values='salary_year_avg', index='job_country', aggfunc='median', columns='job_title_short')
```

```python
top_country
```

```python
df_job_country_salary = df_job_country_salary.loc[top_country]
job_title = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_job_country_salary = df_job_country_salary[job_title]
```

```python
df_job_country_salary
```

```python
df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary ($USD)')
plt.xlabel(' ')
plt.title('Median salary by Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```
