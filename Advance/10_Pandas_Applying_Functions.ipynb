```python
def inflation(salary):
  return salary * 1.03

df_salary['inflatioj_salary']= df_salary['salary_year_avg'].apply(inflation)
#un altro modo è usare lambda:
# df['inflation_salary] = df['salary_year_avg].aply(lambda salary: salary * 1.03)
```

```python
import ast
```

```python
def clean_list(skill_list):
  if pd.notna(skill_list):
    return ast.literal_eval(skill_list)
  
#oppure si può scrivere usando lambda:

#df['job_skills'] = df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)

df['job_skills'] = df['job_skills'].apply(clean_list)
```

```python
def role_inflation(row):
  if "Senior" in row['job_title_short']:
    return 1.05 * row['salary_year_avg']
  else:
    return 1.03 * row['salary_year_avg']
  
df_1 = df.copy()
df_1 = df_1[pd.notna(df_1['salary_year_avg'])]

df_1['salary_year_avg_infl'] = df_1.apply(role_inflation, axis=1)
```
