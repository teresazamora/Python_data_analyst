```python
import pandas as pd
import ast
import matplotlib.pyplot as plt

from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()
```

```python
df.job_posted_date = pd.to_datetime(df['job_posted_date'])
```

```python
df.job_skills = df.job_skills.apply(lambda skill: ast.literal_eval(skill) if pd.notna(skill) else skill)
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US.head()
```

```python
df_DA_US = df_DA_US.dropna(subset='salary_year_avg')
```

```python
df_DA_US.sample(10)
```

```python
plt.boxplot(df_DA_US.salary_year_avg, vert=False)
```

```python
job_titles = ['Data Analyst', 'Data Scientist', 'Data Engineer']
```

```python
df_US = df[(df['job_title_short'].isin(job_titles)) & (df['job_country'] == 'United States')].copy()
```

```python
df_US.head()
```

```python
df_US = df_US.dropna(subset='salary_year_avg')
```

```python
df_US.head()
```

```python
df_US = [df_US[df_US['job_title_short'] == job_title]['salary_year_avg'] for job_title in job_titles] #df_US trasforma il data frame in una lista (perch boxplot accetta come il valore solo array o liste) filtrando gli elementi per job_title, ma estraendo solo la colonna di salaty_year_avg quindi la lista sara compresa solo dei valori di salari che corrispondono a ciascuno dei tre mestieri indicati nella lista job_titles

#list comprehension (generatore delle liste) -  una tecnica procedurale per creare liste nel linguaggio a partire da una lista iniziale.
# ad esempio abbiamo una lista a = [1,2,3,4] e vogliamo creare una nuova lista dove gli elementi saranno doppio della lista a:
# risultato atteso b = [2,4,6,8] quindi si puo procedere con il ciclo for:
# b = [] for num in a: b.append(num * 2), oppure si pu procedere con la list comprehension che semplifica la lettura del codice:
# b = [num * 2 for num in a]
```

```python
plt.boxplot(df_US, labels=job_titles, vert=False)
plt.show()
```
