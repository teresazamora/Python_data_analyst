```python
import pandas as pd
import matplotlib.pyplot as plt
import datasets

from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')

df_2 = dataset['train'].to_pandas()

df_2['job_posted_date'] = pd.to_datetime(df_2['job_posted_date'])
```

```python
df_2
```

```python
df_2.info()
```

```python
median_count = df_2.groupby('job_title_short')['salary_year_avg'].agg('median')
median_count = median_count.sort_values()
```

```python
median_count.plot(kind='barh', title='Median Salary by Job Title', xlabel='Salary($USD)',ylabel='')
```
